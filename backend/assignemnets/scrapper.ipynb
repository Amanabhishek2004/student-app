```python

api_key = "AIzaSyD7sJLZXaChqNmn6FY68HRmmYd8TACO_UM"

# Create a YouTube service object
youtube = build('youtube', 'v3', developerKey=api_key)

# Function to get channel ID from video ID
def get_channel_id_from_video(video_id):
    request = youtube.videos().list(
        part='snippet',
        id=video_id
    )
    response = request.execute()
    
    if response['items']:
        # Extract the channel ID from the video's snippet
        return response['items'][0]['snippet']['channelId']
    else:
        return None
```

```python
def get_channel_details(channel_id):
    request = youtube.channels().list(
        part='snippet,contentDetails,statistics',
        id=channel_id
    )
    response = request.execute()
    return response['items'][0]['snippet']
```

```python
df = pd.read_csv("youtube_links1.csv")
df["link"].values
```

```python
def filter_link(string):
    if "https://youtube.com/watch" in string:
        return ""
    else:
        return string



samples = []

for i in df["link"].values:
       if isinstance(i , str):
           samples.append(filter_link(i))
         
samples.remove("")
samples
```

```python
channelslinks = []
for i in df.iterrows():
    link = i[1].get("link").find("=")
    
    if link !=-1:
       channel_id =  get_channel_id_from_video(i[1].get("link")[link+1:]) 
       det = get_channel_details(channel_id)
       channelslinks.append(f"www.youtube.com/{det}")
       
      
```
